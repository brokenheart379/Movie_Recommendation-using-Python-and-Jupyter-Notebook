```python
star_wars_user_ratings.head()
```

```python
similar_to_star_wars=movie_matrix.corrwith(star_wars_user_ratings)
similar_to_star_wars
```

```python
corrleation_to_star_wars=pd.DataFrame(similar_to_star_wars,columns=['Correlation'])
corrleation_to_star_wars
```

# Predict Function

```python
def predict_movies(movie_name):
    movie_user_rating=movie_matrix[movie_name]
    similar_to_movie=movie_matrix.corrwith(movie_user_rating)
    
    corr_movie=pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=False)
    corr_movie=corr_movie.join(ratings['number_of_rating'])
    predictions=corr_movie[corr_movie['number_of_rating']>100].sort_values('Correlation',ascending=False)
    return predictions
```

```python
prediction=predict_movies("Titanic (1997)")
```
